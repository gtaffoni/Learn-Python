## Plotting with more details 

The main idea with _object-oriented_ python is to create objects that one can apply functions and actions on.

A figure in python is an object, working with figure objects become exrtremely powerful in particular when 
complex images with descriptions and subplots must be done. And this is true in any scientific work we approach.

To use the object-oriented approach you will reporduce the same plot as  previous example, but instead of creating a new global figure instance we store a reference to the newly created figure instance in the *Figure* variable, and from it we create a new axis instance axes using the add_axes method in the Figure class instance *Figure*.

Create a new figure of size 8x6 points, using 100 dots per inch

```python
plt.figure(figsize=(8, 6), dpi=100)

plt.subplot(111)
plt.plot(X, C, color="blue", linewidth=1.0, linestyle="-")

# Plot sine using green color with a continuous line of width 1 (pixels)
plt.plot(X, S, color="green", linewidth=1.0, linestyle="-")

# Set x limits
plt.xlim(-4., 4.)

# Set x ticks
plt.xticks(np.linspace(-4, 4, 9, endpoint=True))

# Set y limits
plt.ylim(-1.0, 1.0)

# Set y ticks
plt.yticks(np.linspace(-1, 1, 5, endpoint=True))

```

##  Plots with full control on the design

A “figure” in matplotlib means the whole window in the user interface.  Within this figure there can be “subplots”.
So far we have used implicit figure and axes creation. This is handy for fast plots. 

Explicitely define figures, axes, subplot is a  further step in the use of _object-orientet_ plotting. 
While subplot positions the plots in a regular grid, axes allows free placement within the figure. Both can be useful depending on your intention.  If there is none it calls figure() to make one, strictly speaking, to make a subplot(111). Let’s look at the details.


```python
fig = plt.figure()

axes = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # left, bottom, width, height (range 0 to 1)

axes.plot(X, C, 'r')

axes.set_xlabel('x [Rad]')
axes.set_ylabel('$cos(x)$')
axes.set_title('plotting example');
```

Although a little bit more code is involved, the advantage is that we now have full control of where the plot axes are placed, and we can easily add more than one axis to the figure:

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # main axes
axes2 = fig.add_axes([0.4, 0.25, 0.2, 0.3]) # inset axes

# main figure
axes1.plot(X, C, 'r')
axes1.set_xlabel('x')
axes1.set_ylabel('cos(x)')
axes1.set_title('plot example')

# insert
axes2.plot(X, S, 'g')
axes2.set_xlabel('x')
axes2.set_ylabel('sin(x)')
axes2.set_title('sub plot');
```

### Bautify your plot: changing fonts and axes

**Manipulating axes and thicks**

The appearance of the axes is an important aspect of a figure that we often need to modify to make a publication quality graphics. We need to be able to control where the ticks and labels are placed, modify the font size and possibly the labels used on the axes. In this section we will look at controling those properties in a matplotlib figure.


```python
fig, axes = plt.subplots(1, 3, figsize=(12, 4))

axes[0].plot(X, C)
axes[0].set_title("default axes ranges")

axes[1].plot(X,C)
axes[1].axis('tight')
axes[1].set_title("tight axes")

axes[2].plot(X,C)
axes[2].set_ylim([0, 1])
axes[2].set_xlim([-np.pi, 0])
axes[2].set_title("custom axes range");
```

Logaritmic scale

We can explicitly determine where we want the axis ticks with set_xticks and set_yticks, which both take a list of values for where on the axis the ticks are to be placed.

```python
xt=[-1,-0.5,0,0.5,1] # thick list
```

```python
fig, ax = plt.subplots()
ax.plot(X, C, lw=2)
ax.set_xlim(-1,1)
ax.set_xticks(xt)
ax.set_ylim(0.4,1)
ax.set_xticklabels([r'$\alpha$', r'$\beta$', r'$\gamma$', r'$\delta$', r'$\epsilon$'], fontsize=18)
```

It is often better use scientific notation:

```python
fig, ax = plt.subplots()
ax.plot(X, C, lw=2)
ax.set_xlim(-1,1)
ax.set_ylim(0.5,1)
ax.set_yticks([0.5,0.7, 0.9, 1.])
ax.set_xticks([-1,-0.5,0,0.5,1])
ax.set_yscale("log")

from matplotlib import ticker
formatter = ticker.ScalarFormatter(useMathText=True)
formatter.set_scientific(True) 
formatter.set_powerlimits((-1,1)) 
ax.yaxis.set_major_formatter(formatter) 
```

With the grid method in the axis object, we can turn on and off grid lines. We can also customize the appearance of the grid lines using the same keyword arguments as the plot function:

```python
fig, ax = plt.subplots()
ax.plot(X, C, lw=2)
ax.set_xlim(-1,1)
ax.set_ylim(0.5,1)
ax.set_yticks([0.5,0.7, 0.9, 1.])
ax.set_xticks([-1,-0.5,0,0.5,1])
ax.grid(True)
```

Sometimes it is useful to have dual x or y axes in a figure; for example, when plotting curves with different units together.

```python
fig, ax = plt.subplots()

ax.plot(X, C, lw=2, color="blue")
ax.set_ylabel(r"cos(x)", fontsize=18, color="blue")

ax1 = ax.twinx()
ax1.plot(X, S, lw=2, color="red")
ax1.set_ylabel(r"sin(x)", fontsize=18, color="red")
ax.set_xlabel('x', fontsize=18 )
```

## Histograms

In scientific papers it is often necessary to make histograms from data. Here we show how to make it using a very simple example based on random numbers.

```python
# A histogram
n = np.random.randn(100000)
fig, axes = plt.subplots(1, 2, figsize=(12,4))

axes[0].hist(n)
axes[0].set_title("Default histogram")
axes[0].set_xlim((min(n), max(n)))

axes[1].hist(n, cumulative=True, bins=50)
axes[1].set_title("Cumulative detailed histogram")
axes[1].set_xlim((min(n), max(n)));
```

## Plot from data and saving plots

In science usually, we plot data from files that stores the ouput of data analysis software or simulations.
numpy implements a function to easely read data

Signature: np.loadtxt(fname, dtype=<class 'float'>, comments='#', delimiter=None, converters=None, skiprows=0, usecols=None, unpack=False, ndmin=0)
Docstring:
Load data from a text file.

Each row in the text file must have the same number of values.

```python
DATA=np.loadtxt('BH2015_1x10x4.dat')
```

```python
DATA.shape
```

**Note** the array is loaded in raw order. We need to transpose it.

```python
DATA=np.loadtxt('BH2015_1x10x4.dat').T
```

```python
DATA.shape
```

This file contains the computing time of a N-Body code. The first column is the expantion factor $a(t)$, the second is the total computing time, the third is the gravity force computing time....(more info in the header of the file).

```python
fig, ax = plt.subplots()

ax.plot(DATA[0], DATA[1], lw=2, color="blue")
ax.set_ylabel(r"Computing time [sec]", fontsize=18)
ax.set_xlabel(r"a(t)",fontsize=18)
```

You can save the plot in different formats (png, ps, pdf etc.)

```python
fig, ax = plt.subplots()

ax.plot(DATA[0], DATA[1], lw=2, color="blue")
ax.set_ylabel(r"Computing time [sec]")
ax.set_xlabel(r"a(t)")
plt.savefig("exercise_2.png", dpi=72)
```
